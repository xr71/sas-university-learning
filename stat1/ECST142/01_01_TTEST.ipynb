```sas
libname STAT1 "/folders/myfolders/stat1/ECST142/data/";
```

```sas
proc print data=stat1.normtemp (obs=5);
run;
```

```sas
proc univariate data=stat1.normtemp;
    var bodytemp heartrate;
    histogram bodytemp heartrate / normal kernel;
    inset n mean std / position=ne;
run;
```

```sas
* normal body temperature is 98.6;
* let's test this;

proc ttest data=stat1.normtemp H0=98.6;
    var bodytemp;
run;
```

```sas
* normal body temperature is 98.6;
* let's test this;

proc ttest data=stat1.normtemp H0=98.6;
    var bodytemp;
    class gender;
run;
```

```sas
proc print data=stat1.german (obs=5); run;
```

```sas
proc freq data=stat1.german;
    table Group;
run;
```

```sas
proc means data=stat1.german;
    var change pre retain;
    class group;
run;
```

```sas
proc ttest data=stat1.german;
    class group;
    var change;
run;
```

```sas
ods graphics; 
proc ttest data=STAT1.German plots(shownull)=interval; 
   class Group; 
   var Change; 
   title "German Grammar Training, Comparing Treatment to Control"; 
run;
```
